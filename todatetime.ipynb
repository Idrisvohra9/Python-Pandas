```python
import pandas as pd
import matplotlib.pyplot as plt
# Data to be timeSeries
data = {
    "temperature": [22, 23, 23, 24, 24, 25, 26, 27, 28],
    "humidity": [65, 64, 63, 62, 61, 60, 59, 58, 57],
    "location": ["I", "a","v","q","w","v", "b","l","c"]
}
print(len(data["temperature"]))
# Creating a data frame:
df = pd.DataFrame(data)

# Indexing based on time series:

# Seconds timestamp:
print("Seconds")
df.index = pd.to_datetime(pd.date_range("2023-03-29", periods=9, freq="S"))
display(df)

# Minutely timestamp:
print("Minutes")
df.index = pd.to_datetime(pd.date_range("2023-03-29", periods=9, freq="T"))
display(df)

# Hourly timestamp:
print("Hours")
df.index = pd.to_datetime(pd.date_range("2023-03-29", periods=9, freq="H"))
display(df)

# Daily Timestamp
print("Daily")
df.index = pd.to_datetime(pd.date_range("2023-03-29", periods=9, freq="D"))
display(df)

# Month Timestamp
print("Month")
df.index = pd.to_datetime(pd.date_range("2023-03-29", periods=9, freq="M"))
display(df)

# Year Timestamp
print("Year")
df.index = pd.to_datetime(pd.date_range("2023-03-29", periods=9, freq="Y"))
display(df)


fig, ax = plt.subplots(figsize=(20,10))
df.plot(title="Timestamp for Temperature and Humidity", xlabel="Time", ylabel="Values", color=["#1d1d1b", "#604eab"], ax=ax)
# df.plot()
plt.show()
```
